```python
from dataclasses import dataclass

@dataclass

class ParametrosReator:
  VazaoVolumetrica: float
  VolumeReator: float
  VelocidadeEspecifica: float
```

```python
## modularizando

def CalculadoraConcentracao (CAin,Parametros):

  F = Parametros.VazaoVolumetrica
  V = Parametros.VolumeReator
  k = Parametros.VelocidadeEspecifica

  CA = F / (F + V * k) * CAin
  return CA

```

```python


def SimuladorReator (CAin, Parametros, NumeroReatores):

  SolucaoCA = []
  SolucaoX = []

  for i in range (NumeroReatores):

    CA_novo = CalculadoraConcentracao (CAin,Parametros)

    X = (CA_novo - CAin) / CAin

    SolucaoX.append(X)

    SolucaoCA.append(CA_novo)

    CAin= CA_novo
    return SolucaoCA, SolucaoX



```
