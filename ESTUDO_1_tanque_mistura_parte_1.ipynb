```python
from dataclasses import dataclass

@dataclass

class Corrente:
  Vazao: float
  FracaoMassica: float
```

```python
def Misturador(C1:Corrente,C2:Corrente):

  Fa= C1.Vazao
  Xa = C1.FracaoMassica
  Fb = C2.Vazao
  Xb = C2.FracaoMassica

  F = Fa + Fb
  X = (Xa*Fa + Xb*Fb)/F

  return (F,X)
```
