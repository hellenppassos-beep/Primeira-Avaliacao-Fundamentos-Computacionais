```python
## QUESTÃO 3 LETRA A) PARTE 1
'''
EU NÃO LI O ENUNCIADO CORRETAMENTE E FIZ APENAS RASCUNHOS
PARA DESENVOLVER O CÓDIGO EM VEZ DE REALIZAR O PSEUDOCÓDIGO DE FORMA FORMATADA,
ESPERO QUE ACEITE O MEU RESULTADO FINAL
'''
##CÓDIGO COM FORMALISMOS DE ERRO
##ESSE EU FIZ SEM INPUT PORÉM ELE POSSUI A VALIDAÇÃO DE ERRO PARA STRING

NivelTanque = 5
CapacidadeTanque = 30 ## metros cúbicos


try:
  if not isinstance(NivelTanque, (int,float)):
    raise ValueError('Valor inválido')
  if NivelTanque < 0:
    raise ValueError('Valor inválido')

  if (NivelTanque / CapacidadeTanque) < 0.2:
    print(f'Nível baixo')


  elif (NivelTanque / CapacidadeTanque) > 0.8:
    print(f'Risco de transbordamento')

  else:
    print(f'Operação normal')

except ValueError as e:
  print(f"Falha de: {e}")
```

```python
##QUESTÃO 3 LETRA A) PARTE 2
##COM INPUT E VALIDAÇÃOD E ERRO PARA NEGATIVOS (PARA CARACTERES DIFERENTES DE NÚMEROS NÃO CHEGUEI A UMA CONCLUSÃO DE COMO FAZER)
Contador = 0

while Contador < 3:
  NivelTanque = float(input("Informe o nível atual do tanque: "))

  try:
      if NivelTanque < 0:
        raise ValueError('Valor inválido')

  except ValueError as e:
    print(f"Falha de: {e}")
    continue

  if (NivelTanque / CapacidadeTanque) < 0.2:
    Contador +=  1
    print(f'Nível baixo')

  elif (NivelTanque / CapacidadeTanque) > 0.8:
    print(f'Risco de transbordamento')

  else:
    print(f'Operação normal')

print(f'INTERROMPER PROCESSO!!!')
```

```python
##QUESTÃO 4
import random
import numpy as np

def AnaliseConcentracao(Concentracao):

  '''
  ESSA FUNÇÃO ANALISA A CONCENTRAÇÃO MEDIDA E DIZ SE É BAIXA, ALTA OU ADEQUADA
  '''

  if Concentracao < 3:
    return'Baixa'
  elif Concentracao > 7:
    return'Alta'
  else:
    return'Adequada'

def Contador(Concentracao):

  '''
  ESSA FUNÇÃO CONTA QUANTAS VEZES CADA ANÁLISE ACONTECEU E DEVOLVE UMA LISTA
  EM QUE O PRIMEIRO ELEMENTO É A QUANTIDADE DE VEZES QUE A CONCENTRAÇÃO FOI BAIXA,
  O SEGUNDO ELEMENTO É A QUANTIDADE DE VEZES QUE A CONCENTRAÇÃO FOI ALTA E O
  TERCEIRO ELEMENTO É A QUANTIDADE DE VEZES QUE A CONCENTRAÇÃO FOI ADEQUADA
  '''

  Contador_Baixa,Contador_Alta,Contador_Adequada = 0,0,0
  for i in range (0,10):

    if Concentracao[i] < 3:
      Contador_Baixa += 1

    elif Concentracao[i] > 7:
      Contador_Alta += 1
    else:
      Contador_Adequada += 1
  return [Contador_Baixa,Contador_Alta,Contador_Adequada]

Concentracao = np.empty(10) # Aqui eu criei um vetor vazio com 10 valores quaisquer

# Aqui eu sorteio um número de 0 a 10 por 10 vezes em que a cada sorteio, o valor é alocado no vetor concentração com a formatação de apenas 1 casa decimal

for i in range (10):
  Sorteios = random.uniform(0, 10)
  Concentracao[i] = round(Sorteios, 1)

for i in range (10): # Aqui eu printo a Concentração medida e a análise feita pela função AnaliseConcentracao
  print(f'Concentração = {Concentracao[i]} Análise: {AnaliseConcentracao(Concentracao[i])}')


Resumo = Contador(Concentracao) #Aqui "Resumo" recebe o output da função Contador, que é a lista com as frequências

print(f' \n>>>>>>>> Resumo <<<<<<<< \n\n BAIXA: frequência = {Resumo[0]}\n ADEQUADA: frequência = {Resumo[1]}\n ALTA: frequência = {Resumo[2]}')

# Por último, As frequências são comparadas até que se encontre uma predominante ou um empate

if Resumo[0] > Resumo[1] and Resumo[0] > Resumo[2]:
    print('\n Predominante: Baixa')
elif Resumo[1] > Resumo[0] and Resumo[1] > Resumo[2]:
    print('\n Predominante: Adequada')
elif Resumo[2] > Resumo[0] and Resumo[2] > Resumo[1]:
    print('\n Análise Predominante: Alta')
else:
    print('\n Não houve uma análise predominante')



```
