```python
from dataclasses import dataclass

@dataclass

class Corrente:
  Vazao: float
  FracaoMassica: float

A = Corrente(100,0.5)

```

```python
def Misturador(C1: Corrente, C2: Corrente):
  """
  Calcula a vazão e a frção mássica de uma mistura de correntes
  Entradas:
  C1: Corrente
  C2: Corrente
  Saídas:
  F: Vazão da mistura
  X: Fração mássica da mistura
  """

  Fa = C1.Vazao
  Xa = C1.FracaoMassica

  Fb = C2.Vazao
  Xb = C2.FracaoMassica

  F = Fa + Fb
  X = (Fa*Xa + Fb*Xb)/F

  return (F,X)
```
